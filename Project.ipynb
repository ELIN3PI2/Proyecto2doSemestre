```python
import pandas as pd
import re
import os

current_directory = os.getcwd()

# Crear el path completo al archivo reportes.txt
reports_path = os.path.join(current_directory, "Web scraping", "reportes.txt")
# Abrir el txt y almacenar los reportes del txt en una variable
with open(reports_path, "r") as file:
    reports = file.read()

dates_path=os.path.join(current_directory,"Web scraping","links.csv")
#crear un data frame a partir de la columna de las fechas del csv
with open(dates_path,"r") as f: 
    url=pd.read_csv(f)

data=url.iloc[:, [3]]
#renombrando las columnas
data.columns=['Fecha']
# Convertir la columna de "Fecha" a formato de fecha
data['Fecha'] = pd.to_datetime(data['Fecha'], format='%m %d %Y',errors='coerce')
# Crear una nueva columna con el mes
data['mes'] = data['Fecha'].dt.month


#extraer la informacion referente a la maxima afectacion
def max_afectation():
    #patron a detectar
    keyword= r'\bmáxima afectación \b.*?(\d+).*?\bMW\b ' 
    #busca las coincidencias
    coincidences=re.findall(keyword, reports)
    global data
    data["Máxima afectación"]= (coincidences + [None]*len(data))[:len(data)]
    return data["Máxima afectación"]

#extraer la información referente a la disponibilidad
def availability():
    keyword= r'\bdisponibilidad \b(?:.(?!\b07\b))*?(\d+).*?\bMW\b'
    coincidences=re.findall(keyword,reports)
    global data
    data["Disponibilidad"]= (coincidences+ [None]*len(data))[:len(data)]
    return data["Disponibilidad"]

max_afectation()
availability()
data
```
