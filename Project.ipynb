```python
import pandas as pd
import re
#abrir el txt y almacenar los reportes del txt en una variable
with open("./reportes.txt", "r") as file:
    reports=file.read()
with open("./url.csv","r") as f:
    url=pd.read_csv(f)
data=url.iloc[:, [3]]

#extraer la informacion referente a la maxima afectacion
def max_afectation():
    #patron a detectar
    keyword= r'\bmáxima afectación \b.*?(\d+).*?\bMW\b' 
    #busca las coincidencias
    coincidences=re.findall(keyword, reports)
    global data
    data["Máxima afectación"]= (coincidences + [None]*len(data))[:len(data)]
    return data["Máxima afectación"]

#extraer la información referente a la disponibilidad
def availability():
    keyword= r'\bdisponibilidad \b(?:.(?!\b07\b))*?(\d+).*?\bMW\b'
    coincidences=re.findall(keyword,reports)
    global data
    data["Disponibilidad"]= (coincidences+ [None]*len(data))[:len(data)]
    return data["Disponibilidad"]

max_afectation()
availability()
data
```
