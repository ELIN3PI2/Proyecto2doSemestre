```python
import csv
import requests
from bs4 import BeautifulSoup
import re

# Leer el archivo CSV
with open('./links.csv', 'r') as f:
      reader = csv.reader(f)
      # Obtener las URL que aparecen dentro del fichero en la columna correspondiente
      urls = [row[1] for row in reader]

#obtener los datos de cada una de las urls y parsear el html
responses=[]
for url in urls:
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')
    responses.append(soup)
    
```

```python
#crear una lista vacia que almacene el contenido visible de cada uno de los reportes
text=[]
#obtener el texto visible de cada uno de los urls
for content in responses:
    visible_text= content.get_text()
    text.append(visible_text)
print(text)


```

```python
max_first_word="máxima afectación"
max_last_word="MW"

# Abre el archivo en modo de escritura
file = open('resultado.csv', 'w', newline='')
# Crea un escritor CSV
writer = csv.writer(file)


def max_afectation():
   for texto in text:
       afectation= re.findall(f"{max_first_word}(.*?){max_last_word}", texto)
       for r in afectation:
           writer.writerow([r])

#cargar la información filtrada en el  archivo csv
max_afectation()
# Cierra el archivo
file.close()         

          

       
           
```
